```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    y = 0.65 + 5.68*(np.log(1 + x/5.68)) - x
    return y

x = np.linspace(0, 5, 100)
# 起始值設定
xa = 0.
xb = 10.
i = 1
xc = (xa + xb)/2

rootList = []
iterationList = []
F_diff = []

# 10^-5
while abs(f(xc))>1e-5:
    diff = abs(xb-xa) 
    
    if f(xc) * f(xa) > 0:
        xa = xc
    else:
        xb = xc

    xc = (xa+xb)/2
    i += 1
    print(i, xc)
    iterationList.append(i)
    rootList.append(xc)
    F_diff.append(diff) 


plt.plot(x, f(x))
plt.grid()
plt.show()
```

```python
#繪製F迭代變化過程
plt.figure()
plt.bar(iterationList,rootList)                                    
plt.xlabel('Times')
plt.ylabel('F')
plt.title('Times vs F')
plt.show()
```

```python
#繪製F差異值迭代變化過程
plt.figure()
plt.bar(iterationList,F_diff)
plt.xlabel('Times')
plt.ylabel('Difference')
plt.title('Times vs Difference')
plt.show()
```
